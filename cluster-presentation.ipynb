```python
data1 = data.drop('country',axis=1)
data1.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data1 = scaler.fit_transform(data1)
```

```python
data1 = pd.DataFrame(data1)
data1.columns = data.columns[1:10]
data1.head()
```

```python
#k-means clustering
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
ss = []

for k in range(2,11):
    kmeans = KMeans(n_clusters=k).fit(data1)
    ss.append([k,silhouette_score(data1,kmeans.labels_)])
    
    temp = pd.DataFrame(ss)
    plt.plot(temp[0],temp[1])
```

```python
#elbow curve
ssd = []
for k in range(2,11):
    kmean = KMeans(n_clusters=k).fit(data1)
    ssd.append([k,kmean.inertia_])
temp = pd.DataFrame(ssd)
plt.plot(temp[0],temp[1])
```

```python
kmean = KMeans(n_clusters = 3,random_state=101)
kmean.fit(data1)
```

```python
label = pd.DataFrame(kmean.labels_, columns = ['labels'])
data.kmean = data.copy()
data.kmean = pd.concat([data.kmean,label],axis=1)
data.kmean.head()
```

```python
data.kmean.labels.value_counts()
```

```python
#Scatter plot
plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
sns.scatterplot(x='child_mort', y='income', hue='labels', data=data.kmean, palette='Set1')
plt.title('Child_mort vs Income')

plt.subplot(2,2,2)
sns.scatterplot(x='child_mort', y='gdpp', hue='labels', data=data.kmean, palette='Set1')
plt.title('Child_mort vs GDPP')

plt.subplot(2,2,3)
sns.scatterplot(x='income', y='gdpp', hue='labels', data=data.kmean, palette='Set1')
plt.title('Income vs GDPP')

plt.show()
```

```python
plt.figure(figsize=(15,10))

data.kmean.drop(['country','exports','health','imports','inflation','life_expec','total_fer'],axis=1).groupby('labels').mean().plot(kind='bar')
plt.show()
```
